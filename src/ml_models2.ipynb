```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = pd.read_csv("/content/uspollution_pollution_us_2000_2016.csv",usecols =COL_LIST )
```

ref:https://www.pca.state.mn.us/air-water-land-climate/understanding-the-air-quality-index-aqi#:~:text=The%20AQI%20is%20calculated%20by,National%20Ambient%20Air%20Quality%20Standards.

```python
df1["max_values"] = df1.max(axis=1)
df1
```

```python
def AQI_Range(x):
  if x<=50:
    return "Good"
  elif x>50 and x<=100:return "Moderate"
  elif x>100 and x<=200:return "Poor"
  elif x>200 and x<=300:return "Unhealthy"
  elif x>300 and x<=400:return "Very unhealthy"
  elif x>400:return "Hazardous"
df1['AQI Range'] = df1['max_values'] .apply(AQI_Range)
df1.head()
```

```python
x=df1[['NO2 AQI','O3 AQI','SO2 AQI','CO AQI']]
y=df1[['max_values']]
x.head()
```

## Statistical Test for detecting stationarity:ADF test

```python
from statsmodels.tsa.stattools import adfuller
stat_test = adfuller(df1['CO AQI'])
print('The test statistic for CO AQI is {} and p-value is {}'.format(stat_test[0], stat_test[1]))
```

it is observed that p value of CO AQI is approximately eqaul to 0. it is quite obvious that the time series becomes stationary.

## ACF plot and PACF plot

```python
from statsmodels.tsa.arima.model import ARIMA
import statsmodels.api as sm
from sklearn.metrics import mean_absolute_error, mean_squared_error
```

```python
sm.graphics.tsa.plot_acf(df1['CO AQI'], lags=35);
```

```python
sm.graphics.tsa.plot_pacf(df1['CO AQI'], lags=35);
```

```python
train_len = int(len(df1['CO AQI']) * 0.8)
df_train = df1['CO AQI'][:train_len]
df_test = df1['CO AQI'][train_len:]
start = len(df_train)
end = len(df_train) + len(df_test) - 1
```

```python
ma_model = ARIMA(df_train, order=(0, 0, 5))
ma_result = ma_model.fit()
```

```python
ma_pred = ma_result.predict(start, end)
ma_pred.index = df_test.index
plt.plot(df_train.index, df_train, label='Train')
plt.plot(df_test.index, df_test, label='Test')
plt.plot(ma_pred.index , ma_pred, label='predictions')
plt.legend()
plt.xlabel('Date')
plt.show()

mae_ma = mean_absolute_error(df_test, ma_pred)
mse_ma = mean_squared_error(df_test, ma_pred)
rmse_ma = np.sqrt(mean_squared_error(df_test, ma_pred))
ma_perf = {'mae_ma': mae_ma,
             'mse_ma':mse_ma,
             'rmse_ma':rmse_ma}
ma_perf = pd.DataFrame([ma_perf])
    
print(ma_perf)
```

```python
ar_model = ARIMA(df_train, order=(5, 0, 0))
ar_result = ar_model.fit()
```

```python
ar_pred = ar_result.predict(start, end)
ar_pred.index = df_test.index
plt.plot(df_train.index, df_train, label='Train')
plt.plot(df_test.index, df_test, label='Test')
plt.plot(ar_pred.index , ar_pred, label='predictions')
   
plt.legend()
plt.xlabel('Date')
plt.show()
mae_ar = mean_absolute_error(df_test, ar_pred)
mse_ar = mean_squared_error(df_test, ar_pred)
rmse_ar = np.sqrt(mean_squared_error(df_test, ar_pred))
ar_perf = {'mae_ar': mae_ar,
             'mse_ar':mse_ar,
             'rmse_ar':rmse_ar}
ar_perf = pd.DataFrame([ar_perf])
print(ar_perf)
```

```python
arima_model = ARIMA(df_train, order=(5, 1, 5))
arima_result = arima_model.fit()
```

```python
arima_pred = arima_result.predict(start, end)
arima_pred.index = df_test.index
plt.plot(df_train.index, df_train, label='Train')
plt.plot(df_test.index, df_test, label='Test')
plt.plot(arima_pred.index , arima_pred, label='predictions')

plt.legend()
plt.xlabel('Date')
plt.show()
mae_arima = mean_absolute_error(df_test, arima_pred)
mse_arima = mean_squared_error(df_test, arima_pred)
rmse_arima = np.sqrt(mean_squared_error(df_test, arima_pred))
arima_perf = {'mae_arima': mae_arima,
             'mse_arima':mse_arima,
             'rmse_arima':rmse_arima}
arima_perf = pd.DataFrame([arima_perf])
print(arima_perf)
```
