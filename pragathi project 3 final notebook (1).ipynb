## data cleaning

```python
df3.describe()        # getting the summary of the data
```

```python
df3 = df3[~df3["Long Description"].isnull()].copy()     #removing missing rows of long description
```

```python
df3['Act/CFR Number'] = df3['Act/CFR Number'].fillna(0)   #replacing missing values of ACT/CFR number witho zero
```

```python
#replacing missing values of states,zip,FMD Date with zero
df3['State'] = df3['State'].fillna(0)
df3['Zip'] = df3['Zip'].fillna(0)
df3['FMD-145 Date'] = df3['FMD-145 Date'].fillna(0) 
```

```python
df3.isnull().sum()    # getting the sum of all null values present in the data set after cleaning the data
```

```python
# dropping unwanted columns for data analysis
df3=df3.drop(columns=['Zip','FMD-145 Date','Act/CFR Number'])
```

# Hypothesis

### 1.in which fiscal year most of the inspections taken place

```python
#drawing bar graph for number of inspections occured in particular yeaar
#using group by fun ction to get the common inspection ID from both the data sets and then represting no. of inspections occured in a paricular year in a bar graph
no_of_inspections=df3.groupby(['Fiscal Year'])['Inspection ID'].nunique().reset_index(name='inspections')
no_of_inspections.plot.barh(x='Fiscal Year',y='inspections',title="no.of inspections in particular year",rot=0)
plot.xlabel("no of inspections")
```

#### from the above bar graph it is observed that maximumof inspections occured in the 2011 i.e is approximately 6000 inspections

### 2.in which product type or program area maximum inspections occurred

```python
# using groupby to get common program area in both the data sets .
program_area=df3.groupby(['Program Area'])['Inspection ID'].nunique().reset_index(name='inspections')
program_area.plot.barh(x='Program Area',y='inspections',title="no.of inspections in particular program area",rot=0)
plot.xlabel("no.of inspections")
```

```python
program_area=df3.groupby(['Product Type'])['Inspection ID'].nunique().reset_index(name='inspections')
program_area.plot.barh(x='Product Type',y='inspections',title="no.of inspections in particular product type",rot=0)
plot.xlabel("no.of inspections")
```

#### from the above 2 graphs it is ibserved that maximum number of inspections are occured in food related products

###  3.which type of action indicated mostly.

```python
classification=df3.groupby(['Classification'])['Inspection ID'].nunique().reset_index(name='inspections')
classification.plot.line(x='Classification',y='inspections')
plot.show()
```

```python
df3['Classification'].value_counts(normalize=True).plot.pie()    #drawing pieplot for inspection results
plot.show()
```

#### from the above plot it is observed that maximum action indicated for inspection is voluntary action

### 4.which is the state for least inspections.

```python
states=df3.groupby(['State'])['Inspection ID'].nunique().reset_index(name='inspections')    #getting the no. of inpspections occured in each state
states
```

```python
states.nsmallest(5,'inspections')   #getting the states in which very less inspections occured
```

### state in which least number of inspections occured is American Samoa

## 5.in which country maximum number of inpsections occured

```python
# getting number of inspectoins occured in each state
countries=df3.groupby(['Country/Area'])['Inspection ID'].nunique().reset_index(name='inspections')
countries
```

```python
countries.nlargest(5,'inspections')   # getting the country names in which more inspections occured
```

```python
countries.nsmallest(10,'inspections')    # getting the country name in which less inspections occured 
```

### from the above dataset it is observed that maximum number of inspections occured in united states.
